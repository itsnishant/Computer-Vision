```python
import cv2
from matplotlib import pyplot as plt
```

```python
#Read the image
img=cv2.imread("boy.jpg") #Opencv read images in BGR by default
plt.imshow(img)
plt.show()
```

```python
#Created a Function for image preprocessing.
def image_sharpening(img):
    
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)#Converted BGR image into Gray Scale.
    img1=cv2.resize(img1,(550,550),interpolation=cv2.INTER_CUBIC) #Resize the images 
    img1=cv2.GaussianBlur(img1,(3,3),0) # Remove the noise using GaussianBlur Filter
    img1=cv2.Laplacian(img1,cv2.CV_64F) #Sharpen the edges in Image using Laplacian filter
    return img1
```

```python
#results can be visible in this section
results=image_sharpening(img)
plt.imshow(results)
plt.show()
```
